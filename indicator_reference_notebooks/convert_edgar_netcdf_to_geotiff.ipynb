```python
import xarray as xr
import rioxarray as rio
```

```python
year = 2015
species = {
    'N2O': 'n2o',
    'CH4': 'ch4',
    'CO2_excl_short-cycle_org_C': 'co2',
    'CO2_org_short-cycle_C': 'co2'
}
speciesfilenamepart = 'CH4'
nc_file = xr.open_dataset('v7.0_FT2021_{0}_{1}_TOTALS.0.1x0.1.nc'.format(speciesfilenamepart, year))
x = nc_file['emi_{0}'.format(species[speciesfilenamepart])]
x = x.rio.set_spatial_dims(x_dim='lon', y_dim='lat')
x.rio.write_crs("epsg:4326", inplace=True)
x.rio.to_raster(r"edgar_{0}_{1}.tiff".format(speciesfilenamepart, year))
```

```python
species = {
    'N2O': 'n2o',
    'CH4': 'ch4',
    'CO2_excl_short-cycle_org_C': 'co2',
    'CO2_org_short-cycle_C': 'co2'
}
for speciesfilenamepart in species:
    for year in [2000, 2005, 2010, 2015, 2016, 2017, 2018, 2019, 2020, 2021]:
        nc_file = xr.open_dataset('v7.0_FT2021_{0}_{1}_TOTALS.0.1x0.1/v7.0_FT2021_{0}_{1}_TOTALS.0.1x0.1.nc'.format(speciesfilenamepart, year))
        x = nc_file['emi_{0}'.format(species[speciesfilenamepart])]
        x = x.rio.set_spatial_dims(x_dim='lon', y_dim='lat')
        x.rio.write_crs("epsg:4326", inplace=True)
        x.rio.to_raster(r"edgar_{0}_{1}.tiff".format(speciesfilenamepart, year))
```
